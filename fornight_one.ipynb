```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
df['education'] = encoder.fit_transform(df['education'])
```

```python
df_catgorial = df.select_dtypes(include='object')
```

```python
df_catgorial.columns
```

```python
catgorial = ['workclass', 'marital-status', 'occupation', 'relationship', 'race','sex', 'native-country', 'income']
```

```python
def value_encoder(data,list):
  for i in list:
    data[i] = encoder.fit_transform(data[i])
  return data
```

```python
final_train = value_encoder(df,catgorial)
```

```python
#final_train.head()
```

```python
X = df.drop('hours-per-week',axis=1)
y = df['hours-per-week']
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV,train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=101)
```

```python
scaled = StandardScaler()
```

```python
scaled_X_train = scaled.fit_transform(X_train)
```

```python
scaled_X_test = scaled.transform(X_test) 
```

```python
model = GradientBoostingRegressor()
```

```python
model.fit(scaled_X_train,y_train)
```

```python
def errors(model):
  preds = model.predict(scaled_X_test)
  mae = mean_absolute_error(y_test,preds)
  rmse = np.sqrt(mean_squared_error(y_test,preds))
  return mae,rmse
```

```python
errors(model)
```

```python
gb_model_2 = GradientBoostingRegressor()

# param_grid
param_grid = {
    'n_estimators': [50,100,150,200],
    'learning_rate': [0.01,0.05,0.1],
    'max_depth': [3,4,5,6,9],
    'subsample': [0.6,0.8,0.9],
}

# Grid Search Model
grid_model_1 = GridSearchCV(estimator=gb_model_2, param_grid=param_grid, cv=5, scoring='neg_mean_absolute_error', verbose=3)

# fit the gird model
grid_model_1.fit(scaled_X_train, y_train)
```

```python
grid_model_1.best_params_
```

```python
errors(grid_model_1)
```
